****DATA Cleaning****

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Imputation Method
imputer = SimpleImputer(strategy='mean')

# Applying the imputer to the DataFrame
# Excluding the 'Country' column as it's categorical and has no missing values
GDP = data.copy()
GDP.iloc[:, 1:] = imputer.fit_transform(GDP.iloc[:, 1:])

# Now, GDP is your new DataFrame with missing values imputed
```

****Data Visulization****

```python
import pandas as pd
import plotly.express as px


# Aggregating GDP by year
years = [str(year) for year in range(1980, 2029)]
global_gdp = data[years].sum()

# Creating a DataFrame for the plot
gdp_trends = pd.DataFrame({
    'Year': years,
    'Global GDP': global_gdp
})

# Creating the interactive line plot with customizations
fig = px.line(gdp_trends, x='Year', y='Global GDP', title='Global GDP Trends (1980-2028)')


fig.update_layout({'xaxis_title': 'Year',
    'yaxis_title': 'Global GDP (in Trillions)'})

fig.show()
```
